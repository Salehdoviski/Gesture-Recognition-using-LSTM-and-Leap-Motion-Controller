```python
def preprocess_data(df, columns_to_transform):
    x0 = le.fit_transform(df[columns_to_transform[0]])
    x1 = le.fit_transform(df[columns_to_transform[1]])
    x2 = le.fit_transform(df[columns_to_transform[2]])
    x3 = le.fit_transform(df[columns_to_transform[3]])
    x4 = le.fit_transform(df[columns_to_transform[4]])
    df['Ring_Extended']= x0
    df['Thumb_Extended']= x1
    df['Index_Extended']= x2
    df['Pinky_Extended']= x3
    df['Middle_Extended']= x4
    df=df.drop('Sample_Number',axis='columns')
    df=df.drop('Pinky_Type',axis='columns')
    df=df.drop('Index_Type',axis='columns')
    df=df.drop('Thumb_Type',axis='columns')
    df=df.drop('Ring_Type',axis='columns')
    df=df.drop('Middle_Type',axis='columns')
    df=df.drop('Gesture',axis='columns')
    #pca = PCA(n_components=8)
    #scaled_pca = pca.fit_transform(df)
    return df
```

```python
while True:
    input_data = r"C:\Users\saleh\PycharmProjects\pythonProject\data_to_send.csv"
    data_frame = pd.read_csv(input_data)
    data_pipeline = preprocess_data(df=data_frame, columns_to_transform=['Ring_Extended', 'Thumb_Extended',
                                                                            'Index_Extended', 'Pinky_Extended',
                                                                            'Middle_Extended'])
    data_pipeline = data_pipeline / np.abs(data_pipeline).max()

    # Assuming your data is a NumPy array
    data_array = np.array(data_pipeline)

    # Set the chunk size
    chunk_size = 60

    # Initialize an empty list to store predictions
    all_predictions = []
    
    # Iterate over the data and fill the array in chunks
    for i in range(0, len(data_array), chunk_size):
        chunk = data_array[i:i + chunk_size, :]

        # Check if the array is full (contains 60 samples)
        if len(chunk) == chunk_size:
            predictions = model.predict(np.expand_dims(chunk, axis=0))
            predicted_class = np.argmax(predictions)
            all_predictions.append(predicted_class)
        # Print the result once per iteration
        if all_predictions:
            predicted_class = max(set(all_predictions), key=all_predictions.count)

            if predicted_class == 0:
                print("no")
            elif predicted_class == 1:
                print("thanks")
            elif predicted_class == 2:
                print("hello")
            elif predicted_class == 3:
                print("boy")
            elif predicted_class == 4:
                print("wrong")
            elif predicted_class == 5:
                print("cat")
            elif predicted_class == 6:
                print("milk")     
```
