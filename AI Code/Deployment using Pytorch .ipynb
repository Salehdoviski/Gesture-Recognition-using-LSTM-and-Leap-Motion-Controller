```python
def convert_labels_to_numpy(labels):

    # Checking if the length of the labels column is divisible by 60
    if len(labels) % 60 != 0:
        raise ValueError("Length of labels column should be divisible by 60.")

    # Reshaping the labels column into chunks of size 60
    chunks = np.array(labels).reshape(-1, 60)

    # Taking the first element from each chunk to form the final labels array
    final_labels = chunks[:, 0]

    # Reshaping the final labels array to have shape (x, 1)
    final_labels = final_labels.reshape(-1, 1)

    return final_labels
```

```python
import torch
from torch import nn
from torch import optim
```

```python
class LeepModel(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.lstm1 = nn.LSTM(23, 128, num_layers=2, dropout=0.2)
        self.lstm2 = nn.LSTM(128, 128, num_layers=2, dropout=0.2)
        
        self.linears = nn.Sequential(
            nn.Linear(128, 128),
            nn.Dropout(0.25),
            nn.Tanh(),
            nn.Linear(128, 64),
            nn.Dropout(0.25),
            nn.Tanh(),
            nn.Linear(64, 32),
            nn.Dropout(0.25),
            nn.Tanh()
        )
        
        self.final = nn.Linear(32, 7)
        
    def forward(self, x):
        x, _ = self.lstm1(x)
        x, _ = self.lstm2(x)
        x = self.linears(x)
        x = self.final(x)
        return x
```

```python
from torch.utils.data import DataLoader, Dataset

class CustomDataset(Dataset):
    def __init__(self, inputs, labels):
        self.inputs = inputs
        self.labels = labels

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return self.inputs[idx], self.labels[idx]
```

```python
train_dataset = CustomDataset(X_train, y_train)
val_dataset = CustomDataset(X_val, y_val)
test_dataset = CustomDataset(X_test, y_test)
```

```python
import time

def train(model_, epochs, train_loader, test_loader, optimizer, loss_function):
    for epoch in range(epochs):
        start_time = time.time()
        model_.train()

        train_loss = 0
        train_correct = 0
        total = 0

        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch, y_batch

            y_logits = model_(X_batch)
            y_pred = torch.softmax(y_logits, dim=1).argmax(dim=1)

            loss = loss_function(y_logits, y_batch)
            acc = accuracy_score(y_true=y_batch.cpu(), y_pred=y_pred.cpu())

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            train_correct += (y_pred == y_batch).sum().item()
            total += y_batch.size(0)

        train_loss /= len(train_loader)
        train_acc = 100 * train_correct / total
        scheduler.step()
        end_time = time.time()
        epoch_duration = end_time - start_time

        model_.eval()
        test_loss = 0
        test_correct = 0
        total = 0

        with torch.no_grad():
            for X_batch, y_batch in test_loader:
                X_batch, y_batch = X_batch, y_batch
                test_logits = model_(X_batch)
                test_pred = torch.softmax(test_logits, dim=1).argmax(dim=1)

                loss = loss_function(test_logits, y_batch)
                acc = accuracy_score(y_true=y_batch.cpu(), y_pred=test_pred.cpu())

                test_loss += loss.item()
                test_correct += (test_pred == y_batch).sum().item()
                total += y_batch.size(0)

        test_loss /= len(test_loader)
        test_acc = 100 * test_correct / total

        if epoch % 1 == 0:
            print(f"Epoch: {epoch} | Loss: {train_loss:.5f}, Acc: {train_acc:.2f}% | Test Loss: {test_loss:.5f}, Test Acc: {test_acc:.2f}%")
```
